```python
def get_top_words(df, top_num):
    '''Input: dataframe to get word occurrence, top number of occurrence
       Output: dataframe with only top occurring words'''
    df = pd.DataFrame(df.sum().sort_values(ascending=False).head(top_num))
    df = df.reset_index()
    df.columns = ['words', 'counts']
    return df
```

```python
def barplot_words(df, title, word_type):
    '''Input: Dataframe, Title of barplot, type of word
       Output: barplot of top occurring word type'''
    plt.figure(figsize=(12,8))
    plt.title(title, fontsize = 20)
    sns.set_style('whitegrid')
    sns.barplot(data = df, x ='counts', y = 'words', orient = 'h')
    plt.xlabel('Counts', fontsize = 15)
    plt.ylabel(word_type, fontsize = 15)
    plt.tick_params(labelsize = 12)
```

## Exploring target variable `salary_average` <a class="anchor" id="salary"></a> 
---
[Back to top!](#toc)
* Outliers were removed i.e. `salary_average` of values below 1,000 and above 20,0000. `salary_average` below 1,000 are removed as those are unrealistic figures that could possibly be used as placeholders, to be negotiated during the job interview process. `salary_average` above 20,000 are also removed as there are only 28 job listings out of the 4,076 that is offering, suggesting that they may unnecessarily skew the dataset.
* `average_salary` is positively skewed as seen in the histogram and boxplots. Hence, a logarithmic function is applied to make the data more normally distributed.

```python
jobs.salary_average.describe()
```

```python
# removing jobs with outlier salary values
jobs = jobs[jobs['salary_average'] >= 1000]
jobs = jobs[jobs['salary_average'] <= 20000]
jobs.salary_average.describe()
```
